```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Criando um numpy array contendo 100 numeros igualmente espa√ßados entre zero e 2
x = np.linspace(start=0, stop=2, num=100)
```

## Criando somente um AXES:

```python
##################### Criando, ao mesmo tempo, uma figura com um axes dentro dela ##############
# OBS: Aqui n√£o h√° como mudar o tamanho da figura 
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label', fontsize=12)  # Add an x-label to the axes.
ax.set_ylabel('y label', fontsize=16)  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
##################### Criando primeiro a figura e s√≥ depois adicionando um axes dentro dela ##############
# OBS: Aqui h√° como mudar o tamanho da figura 
fig = plt.figure(figsize=(16, 6), facecolor='yellow') 
ax = fig.add_subplot(1,1,1)

plt.setp(ax.get_xticklabels(), rotation=0, fontsize=16, color='red')
plt.setp(ax.get_yticklabels(), rotation=0, fontsize=12, color='brown')

ax.plot(x, x, label='linear') 
ax.plot(x, x**2, label='quadratic') 
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('Total', fontsize=12, color='green')
ax.set_ylabel('Company', fontsize=15, color='blue')
ax.set_title('Company Revenue', fontsize=18, color='purple')
ax.set(xlim=[-0.5, 2.5], ylim=[-2, 10], facecolor='pink')


ax.legend(loc='best', ncol=2, fontsize=12, fancybox='True', shadow='True', title='Legend', 
          title_fontsize=14, facecolor='cyan', edgecolor='green')
ax.grid(b='True', which='both', axis='both', color='blue', linestyle='--', linewidth=0.5)
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.25))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.5))
```

## Criando mais de um AXES:

```python
np.random.seed(19680801)
data = np.random.randn(2, 100)
```

```python
##################### Criando, ao mesmo tempo, uma figura com um axes e SETANDO O TAMANHO DA FIGURA ########
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])
```

```python
fig, axes = plt.subplots(2, 2, sharex=False, sharey=True,
                         figsize=(12, 12))  # Criando uma figura e dentro dela criando 2 AXES
axes[0,1].plot(x, x, label='linear')  # Plot some data on the first axes
axes[0,1].plot(x, x**2, label='quadratic')  # Plot more data on the first axes...
axes[0,1].plot(x, x**3, label='cubic')  # ... and some more.
axes[0,1].set_xlabel('x label')  # Add an x-label to the first axes.
axes[0,1].set_ylabel('y label')  # Add a y-label to the first axes.
axes[0,1].set_title("Simple Plot")  # Add a title to first the axes.
axes[0,1].legend()  # Add a legend to the first axes.

for ax in fig.axes:
    ax.grid(True)
    for label in ax.get_xticklabels():
        label.set_color('green')
    for label in ax.get_yticklabels():
        label.set_color('blue')
    
```

```python
#### Here is an example, not recommended for its beauty, which customizes the axes and tick properties:
# plt.figure creates a matplotlib.figure.Figure instance
fig = plt.figure()
rect = fig.patch  # a rectangle instance
rect.set_facecolor('lightgoldenrodyellow')

ax1 = fig.add_axes([0.1, 0.3, 0.4, 0.4])
rect = ax1.patch
rect.set_facecolor('lightslategray')


for label in ax1.xaxis.get_ticklabels():
    # label is a Text instance
    label.set_color('red')
    label.set_rotation(45)
    label.set_fontsize(16)

for line in ax1.yaxis.get_ticklines():
    # line is a Line2D instance
    line.set_color('green')
    line.set_markersize(25)
    line.set_markeredgewidth(3)

plt.show()
```
