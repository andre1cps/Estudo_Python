```python
## Pacotes para abrir arquivos em pastas:
import os
import glob

## Pacotes para arrumar os dados:
import xarray as xr
import numpy as np
```

```python
## Caminhos para os arquivos:
pasta_raiz = '/mnt/HD_500GB_WD/01_Dados_ARM/Estado_Atmosferico/maopblhtsonde1mcfarlM1_geral/'
horas_sondas = ['05','11','17','23']
path_nc = sorted(glob.glob(pasta_raiz+'maopblhtsonde1mcfarlM1_'+horas_sondas[0]+'/*.cdf'))
```

```python
## Lendo o nome de cada nc, um de cada vez:
for n in range(len(path_nc)):
    arq_n = path_nc[n].split('/')[-1]
    print('Arq. {0}: {1}'.format(n,arq_n))
```

```python
meses_hef_05 = {"jan": [], "fev": [], "mar": [], "abr": [], "mai": [], "jun": [], "jul": [], "ago": [],
                "set": [], "out": [], "nov": [], "dez": []}

# Abrindo os nc, um de cada vez:
for file in path_nc:
    nc = xr.open_dataset(file)
    mes = str(int(nc['time.month'].values[0]))
    pblh_hef = nc['pbl_height_heffter'].where((nc['qc_pbl_height_heffter']==0), drop=False)
    pblh_liu = nc['pbl_height_liu_liang'].where((nc['qc_pbl_height_liu_liang']==0), drop=False)
    if mes == '1':
        meses_hef_05['jan'].append(float(pblh_hef.values))
    else:
        pass
    nc.close()
```
