```python
# Definindo uma f(x):
f = Lambda(x, (x**3 - 3*x + 2)*exp(-x/4) - 1)
f
```

```python
# Definindo a primeira derivada da f anterior:
fl = Lambda(x, diff(f(x),x))
fl
```

```python
# Aqui, vamos ver como computar a reta tangente ao gráfico da função f no ponto x0=−1/2. Lembremos que tal 
# reta tangente tem equação y = f′(x0)(x−x0)+f(x0):
x0 = -1/2
r = Lambda(x, fl(x0)*(x-x0) + f(x0))
r
```

```python
# Graficos de f e da reta tangente:
%matplotlib inline
p = plot(f(x), (x,-2,2), line_color='blue', show=False)
q = plot(r(x), (x,-2,2), line_color='red', show=False)
p.extend(q)
p.show()
```
